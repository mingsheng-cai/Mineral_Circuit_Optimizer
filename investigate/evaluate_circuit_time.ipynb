```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the data
data = pd.read_csv('execution_times.txt')

# Extract data
vector_size = data['VectorSize']
execution_time = data['ExecutionTime(s)']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(vector_size, execution_time, marker='o', linestyle='-', color='b', label='Execution Time')

# Set log scale for both axes
plt.xscale('log')
plt.yscale('log')

# Add a reference line (slope=1)
x_ref = np.logspace(np.log10(min(vector_size)), np.log10(max(vector_size)), num=100)
y_ref = x_ref * execution_time.iloc[0] / vector_size.iloc[0]
plt.plot(x_ref, y_ref, 'r--', label='Reference Slope (slope=1)')

# Add title and labels
plt.title('Execution Time vs. Vector Size (Log-Log Scale)')
plt.xlabel('Vector Size (log scale)')
plt.ylabel('Execution Time (s, log scale)')

# Show legend
plt.legend()

# Show grid
plt.grid(True, which="both", ls="--")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the file
data = pd.read_csv('timings_thread.log')

# Extract data
threads = data['Threads']
time_ms = data['Time(ms)']
efficiency = data['Efficiency']

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting Time (ms) on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Number of Threads')
ax1.set_ylabel('Time (ms)', color=color)
ax1.plot(threads, time_ms, marker='o', color=color, label='Time (ms)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xscale('linear')
ax1.set_yscale('linear')

# Plotting Efficiency on the secondary y-axis
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Efficiency', color=color)  # we already handled the x-label with ax1
ax2.plot(threads, efficiency, marker='x', color=color, label='Efficiency')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yscale('linear')

# Adding title and labels
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Execution Time and Efficiency vs. Number of Threads')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

# Adding grid
plt.grid(True, which="both", ls="--")

# Display the plot
plt.show()
```
