### Visualisation of Circuit

```python
import subprocess
import matplotlib.pyplot as plt
import os
```

```python
# To input the arguments for the script and run from this notebook
# Define the arguments
vector = "[0, 1, 3, 2, 4, 4, 3, 1, 3, 6, 1, 1, 0, 5, 1, 1]"
perf = 167.38
recovery = 20.28
grade = 97.72

# Run the script with the arguments
subprocess.run(['python', 'visualize.py', vector, str(perf), str(recovery), str(grade)])
```

```python
# Display the image
image_path = './circuit_diagram/circuit.png'
if os.path.exists(image_path):
    img = plt.imread(image_path)
    plt.figure(figsize=(10, 8))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
else:
    print("Image not found. Please check the path and script execution.")
```
